```python
nadia=cv2.imread("../DATA/Nadia_Murad.jpg",0)
dennis=cv2.imread("../DATA/Denis_Mukwege.jpg",0)
solvay=cv2.imread("../DATA/solvay_conference.jpg",0)
```

```python
face_cascade=cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml') 
```

```python
def detect_face(img):
    img_c=img.copy()
    
    face_rects=face_cascade.detectMultiScale(img_c)
    
    print(face_rects)
    
    for x,y,w,h in face_rects:
        cv2.rectangle(img_c,(x,y),(x+w,y+h),255,7)
        
    return img_c
```

```python
result=detect_face(dennis)
plt.imshow(result,cmap='gray')
```

```python
result=detect_face(nadia)
plt.imshow(result,cmap='gray')
```

```python
result=detect_face(solvay)
plt.imshow(result,cmap='gray') # multiple printed values shows x,y,w,h or rectangle for multiples faces
```

```python
#Adding few parameters in function for better results
```

```python
def adj_detect_face(img):
    img_c=img.copy()
    
    face_rects=face_cascade.detectMultiScale(img_c,scaleFactor=1.2,minNeighbors=5)
    
    #print(face_rects)
    
    for x,y,w,h in face_rects:
        cv2.rectangle(img_c,(x,y),(x+w,y+h),255,7)
        
    return img_c
```

```python
result=adj_detect_face(solvay)
plt.imshow(result,cmap='gray')
```

```python
eye_cascade=cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_eye.xml') 
```

```python
def detect_eyes(img):
    img_c=img.copy()
    
    eye_rects=eye_cascade.detectMultiScale(img_c,scaleFactor=1.2,minNeighbors=5)
    
    for x,y,w,h in eye_rects:
        cv2.rectangle(img_c,(x,y),(x+w,y+h),255,7)
        
    return img_c
```

```python
result=detect_eyes(nadia)
plt.imshow(result,cmap='gray')
```

```python
result=detect_eyes(dennis)
plt.imshow(result,cmap='gray')
```

```python
result=detect_eyes(solvay)
plt.imshow(result,cmap='gray')
```

```python
cap=cv2.VideoCapture(0)

while cap.isOpened():
    
    ret,frame=cap.read()
    
    #Operations
    
    frame_result=adj_detect_face(frame)
    
    
    #display
    
    cv2.imshow("Face Detection",frame_result)
    
    k= cv2.waitKey(1)
    
    if k == ord('q'):
        break
        
cv2.destroyAllWindows()
```
