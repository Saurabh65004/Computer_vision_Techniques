```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
dog_face=cv2.imread("../DATA/sammy_face.jpg")
```

```python
dog_face=cv2.cvtColor(dog_face,cv2.COLOR_BGR2RGB)
```

```python
dog_face_gray=cv2.cvtColor(dog_face,cv2.COLOR_RGB2GRAY)
```

```python
edges=cv2.Canny(image=dog_face,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
edges=cv2.Canny(image=dog_face,threshold1=200,threshold2=255)
plt.imshow(edges)
```

```python
edges=cv2.Canny(image=dog_face,threshold1=225,threshold2=255)
plt.imshow(edges)
```

```python
median=np.median(dog_face)
median
```

```python
#Lower threshold to be either 0 or 70% of the median value, whichever is greater 
lower=int(max(0,0.7*median))
#Upper threshold to be either 130% of the median value or 255 whichever is smaller
upper=int(min(1.3*median,255))
```

```python
lower
```

```python
upper
```

```python
edges=cv2.Canny(image=dog_face,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
# we may have to test and alter values for better results
edges=cv2.Canny(image=dog_face,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```

```python
#But as the image is having too many edges we should first blur it for better results
```

```python
blur_dog=cv2.blur(dog_face,(5,5))
```

```python
plt.imshow(blur_dog)
```

```python
#blurring produced considerable change, plus altering values may lead to much better results
edges=cv2.Canny(blur_dog,lower,upper)
plt.imshow(edges)
```

```python
#Best result till now
edges=cv2.Canny(blur_dog,lower,upper+50)
plt.imshow(edges)
```

```python
nadia=cv2.imread("../DATA/Nadia_Murad.jpg",0)
```

```python
median=np.median(nadia)
print(median)
lower=int(max(0,0.7*median))
upper=int(min(1.3*median,255))
print(lower)
print(upper)
```

```python
edges=cv2.Canny(nadia,lower,upper)
plt.imshow(edges)
```

```python
edges=cv2.Canny(nadia,lower,upper+50)
plt.imshow(edges)
```

```python
nadia_blur=cv2.blur(nadia,(3,3))
plt.imshow(nadia_blur,cmap="gray")
```

```python
edges_nadia=cv2.Canny(nadia_blur,lower,upper)
plt.imshow(edges_nadia,cmap='gray')
```

```python
sobelx=cv2.Sobel(nadia,cv2.CV_64F,1,0)
plt.subplot(121)
plt.imshow(sobelx,cmap="gray")
sobely=cv2.Sobel(nadia,cv2.CV_64F,0,1)
plt.subplot(122)
plt.imshow(sobely,cmap="gray")
```

```python
blend=cv2.addWeighted(sobelx,0.5,sobely,0.5,2)
plt.imshow(blend,cmap='gray')
```
