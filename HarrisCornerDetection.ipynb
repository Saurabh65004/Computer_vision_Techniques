```python
flat_chess=cv2.imread("../DATA/flat_chessboard.png")
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
real_chess=cv2.imread("../DATA/real_chessboard.jpg")
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
real_chess_gray=cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
real_chess_gray=np.float32(real_chess_gray)
real_chess_gray_=real_chess_gray.copy()
real_chess_=real_chess.copy()
```

```python
flat_chess_gray=cv2.cvtColor(flat_chess,cv2.COLOR_RGB2GRAY)
flat_chess_gray=np.float32(flat_chess_gray)
```

```python
flat_chess_gray_=flat_chess_gray.copy()
flat_chess_=flat_chess.copy()
```

```python
dst=cv2.cornerHarris(src=flat_chess_gray_,blockSize=2,ksize=5,k=0.04)
```

```python
dst=cv2.dilate(dst,None)
```

```python
flat_chess_[dst > 0.01 * dst.max()]=[255,0,0]
```

```python
plt.subplot(121)
plt.imshow(check)

plt.subplot(122)
plt.imshow(flat_chess_)
plt.title("ksize=5")
```

```python
for i in [3,5,7,9]:
    dst=cv2.cornerHarris(src=flat_chess_gray_,blockSize=2,ksize=i,k=0.04)
    dst=cv2.dilate(dst,None)
    flat_chess_[dst > 0.01 * dst.max()]=[255,0,0]
    
    plt.subplot(121)
    plt.imshow(check)

    plt.subplot(122)
    plt.imshow(flat_chess_)
    plt.title("ksize="+str(i))
    
    plt.show()
    print("\n \n")
```

```python
#Increase thresh
```

```python
for i in [0.025,0.01,0.05,0.10,0.50]:
    
    dst=cv2.cornerHarris(src=real_chess_gray_,blockSize=2,ksize=3,k=0.04)
    dst=cv2.dilate(dst,None)
    thresh=i*dst.max()
    real_chess_[dst > thresh]=[255,0,0]
    
    plt.subplot(121)
    plt.imshow(check_real)

    plt.subplot(122)
    plt.imshow(real_chess_)
    plt.title("thresh="+str(i))
    
    plt.show()
    print("\n \n")
```
