```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def load_img():
    img=cv2.imread("../DATA/bricks.jpg").astype(np.float32)/255 
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
 def display_img(img):
    fig=plt.figure(figsize=(8,7))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
i =load_img()
display_img(i)
```

```python
#Load image
```

```python
gamma=1/4   # less than 1
```

```python
result=np.power(i,gamma)
display_img(result)
```

```python
gamma=2                   #gamma greater than 1 , makes image darker
result=np.power(i,gamma)
display_img(result)
```
