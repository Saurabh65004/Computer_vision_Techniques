```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
def display_img(img,cmap=None):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap)
    
```

```python
giraffes=cv2.imread("DATA/giraffes.jpg")
giraffes_=cv2.cvtColor(giraffes,cv2.COLOR_BGR2RGB)
plt.imshow(giraffes_)
```

```python
#Binary threshold
```

```python
giraffes_gray=cv2.cvtColor(giraffes_,cv2.COLOR_RGB2GRAY)
```

```python
ret,thresh1=cv2.threshold(giraffes_gray,127,255,cv2.THRESH_BINARY)
ret
```

```python
plt.imshow(thresh1,cmap="gray")
```

```python
#Converting to HSV
```

```python
giraffes_hsv=cv2.cvtColor(giraffes,cv2.COLOR_BGR2HSV)
```

```python
display_img(giraffes_hsv)
```
