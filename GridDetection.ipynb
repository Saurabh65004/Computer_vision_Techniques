```python
def chess_grid(img,patternSize):
    found,corners=cv2.findChessboardCorners(img,patternSize)
    if found:
        out=cv2.drawChessboardCorners(img,patternSize,corners,found)
        plt.imshow(out)
    else:
        print("No Grids Found")
```

```python
def circles_grid(img,pSize):
    found,corners=cv2.findCirclesGrid(image=img,patternSize=pSize)
    if found:
        out=cv2.drawChessboardCorners(img,pSize,corners,found)
        plt.imshow(out)
    else:
        print("No Grids Found")
```

```python
chess_grid(flat_chess,(7,7))
```

```python
chess_grid(flat_chess,(6,6))
```

```python
chess_grid(flat_chess,(5,5))
```

```python
chess_grid(dots,(10,10))
```
