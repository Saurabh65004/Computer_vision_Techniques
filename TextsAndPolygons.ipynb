```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
blank_img=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img)
```

```python
font=cv2.FONT_HERSHEY_COMPLEX
```

```python
#cv2.putText() method is used to put the text on an image
```

```python
cv2.putText(blank_img,text="Hello Bitch!",org=(10,400),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img,text="hello",org=(10,320),fontFace=font,fontScale=4,color=(255,0,0),thickness=4,lineType=cv2.LINE_4)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img,text="hello",org=(10,220),fontFace=font,fontScale=4,color=(255,0,0),thickness=4,lineType=cv2.LINE_8)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img,text="hello",org=(10,220),fontFace=font,fontScale=4,color=(255,0,0),thickness=4)
plt.imshow(blank_img)
```

```python
vertices=np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
vertices.shape
```

```python
pts=vertices.reshape((-1,1,2))
pts
```

```python
pts.shape
```

```python
#cv2.polyLines() is used to draw the polygon on the given image,in this we have to pass the list of points to be drawn. 

cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=4)
plt.imshow(blank_img)
```
