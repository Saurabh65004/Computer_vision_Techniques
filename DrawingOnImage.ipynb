```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
dog_ori=cv2.imread('../DATA/dog_backpack.jpg')
type(dog_ori)
dog=cv2.cvtColor(dog_ori,cv2.COLOR_BGR2RGB)
plt.imshow(dog)
```

```python
flip=cv2.flip(dog,0)
plt.imshow(flip)
```

```python
cv2.rectangle(dog,(180,410),(625,710),(255,0,0),5)
plt.imshow(dog)
```

```python
vertices=np.array([[375,400],[625,710],[250,740]],dtype=np.int32)
pts=vertices.reshape((-1,1,2))

cv2.polylines(dog,[pts],isClosed=True,color=(255,0,0),thickness=4)
cv2.fillPoly(dog,[pts],(255,0,0))
plt.imshow(dog)
```

```python
def draw(event,x,y,flags,params):
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(dog,(x,y),100,(0,0,255),3)
    

cv2.namedWindow(winname="dogHere")
cv2.setMouseCallback("dogHere",draw)
dog=cv2.imread("dog_backpack.jpg")

while True:
    cv2.imshow("dogHere",dog)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
