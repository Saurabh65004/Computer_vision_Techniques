```python
import numpy as np
from numpy import genfromtxt
```

```python
data=genfromtxt('../DATA/bank_note_data.txt',delimiter=',')
```

```python
data
```

```python
labels=data[:,4]
```

```python
labels   #
```

```python
features=data[:,0:4]
```

```python
features  #
```

```python
#splitting our data into training and test set
```

```python
X= features
y=labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
len(X)
```

```python
len(X_train) #67% of data for training set
```

```python
len(X_test)  #33% for test 
```

```python
#Forcing all our feature data to fall within certain range
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_object= MinMaxScaler()
```

```python
scaler_object.fit(X_train)
```

```python
scaled_X_train=scaler_object.transform(X_train)
```

```python
scaled_X_test=scaler_object.transform(X_test)
```

```python
scaled_X_train.max()
```

```python
scaled_X_train.min()
```

```python
#Building a simple network with keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(4, input_dim=4, activation='relu'))

model.add(Dense(8,activation='relu'))

model.add(Dense(1, activation='sigmoid'))
```

```python
#metrics for model
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(scaled_X_train, y_train, epochs=50, verbose=2)
```

```python
#Model is trained Now we'll test
```

```python
#model.predict_classes(scaled_X_test)
```

```python
model.metrics_names
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predictions=model.predict_classes(scaled_X_test)
```

```python
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```

```python
model.save('bankmodel.h5')
```
